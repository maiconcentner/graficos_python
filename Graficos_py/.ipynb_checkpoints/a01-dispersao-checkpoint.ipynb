# Gráfico de dispersão

```python
"""
- Importando nossa biblioteca:
impor matplotlib.pylot as plt

- Criando o gráfico:
Para o gráfico com pontos, usamos o comando plt.scatter(x , y)
Para gráfico com linha suave, usamos o código plt.plot(x, y)

- Criando legendas dos eixos:
plt.xlabel("String")
plt.ylabel("String")

- Criando um título:
plt.title("String")

- Para a impressão
plt.show()

"""
```

```python
import matplotlib.pyplot as plt
```

```python
# definindo uma função
def funcao_tec_grau(x, a=1, b=1, c=1, d=1):
    y = a*x**3 + b*x**2 + c*x + d
    return y
```

```python
# criação do gráfico
lista_x = range(-5, 5)
lista_y3 = []

for i in lista_x:
    lista_y3.append(funcao_tec_grau(i))
    
plt.plot(lista_x, lista_y3, label = '3º grau', color = 'red') #plot da função
plt.legend() #esse comando é para a legenda aparecer (label)
plt.scatter(lista_x, lista_y3, color = 'red', marker = '*') #comando para inserir os pontos
plt.xlabel("Valores de x") #esse é o nome do eixo x
plt.ylabel("Valores de y") #nome do eixo y
plt.title("Gráfico da função y") #comando para inserir o título
plt.show #exibir o gráfico
```

# Inserindo mais de um gráfico no mesmo plot

```python
# definindo uma segunda função
def funcao_sec_grau(x, a=1, b=0, c=0):
    y = a*x**2 + b*x + c
    return y
```

```python
# criação do gráfico
lista_x = range(-10,10)

lista_y3 = []
lista_y2 = []

for i in lista_x:
    lista_y3.append(funcao_tec_grau(i))
    lista_y2.append(funcao_sec_grau(i))
    
plt.plot(lista_x, lista_y3, label = '3º grau', color = 'red') #plot da função do 3 grau
plt.plot(lista_x, lista_y2, label = '2º grau', color = 'black')

plt.legend() #esse comando é para a legenda aparecer (label)
#plt.scatter(lista_x, lista_y, color = 'red', marker = '*') #comando para inserir os pontos
plt.xlabel("Valores de x") #esse é o nome do eixo x
plt.ylabel("Valores de y") #nome do eixo y
plt.title("Gráfico da função y") #comando para inserir o título
plt.show #exibir o gráfico
```

# Utilizando um outro método (função map)

```python
lista_map = list(map(funcao_sec_grau, lista_x)) #transformando o resultado em uma lista

plt.plot(lista_x, lista_map, label = '2º grau', color = 'purple')
plt.xlabel('Valores de x')
plt.ylabel('Valores de y')
plt.title('Função do 2º grau')
plt.legend()
plt.show()
```
